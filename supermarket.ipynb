```python
int_c = obj_c = 0
for i in df.dtypes:
    if i == 'object':
        obj_c +=1
    else:
        int_c +=1
        
print("Total columns with Numerical data type is ",int_c)
print("Total columns with object data type is ",obj_c)
```
